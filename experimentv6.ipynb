```python
import tensorflow as tf
import cv2
import numpy as np
import glob
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import math
import csv
```

```python
# function to load image as np array
def load_image(addr):
    img = cv2.imread(addr, 0) 
    img = cv2.resize(img, (224,224))
    img = img.astype(np.int32)
    return img
```

```python
def score(eye):
    return np.sum(eye)
```

```python
# returns sum of scores of both eyes of a single result
def total_score(result):
    
    i,j,k,l = result[0][0], result[0][1], result[1][0], result[1][1]
    
    # convert 256 grayscale to 4 grayscale
    eye_left = img2[i:i+25, j:j+25]
    eye_left = eye_left//64
    
    eye_right = img2[k:k+25, l:l+25]
    eye_right = eye_right//64
    
    return score(eye_left) + score(eye_right)
```
